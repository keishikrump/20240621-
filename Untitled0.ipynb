```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.hp_filter import hpfilter

pwt = pd.read_csv('/content/pwt1001.csv')

countries = ['India', 'Japan']
years = range(1990, 2020)
df = pwt[(pwt['country'].isin(countries)) & (pwt['year'].isin(years))]

df_pivot = df.pivot(index='year', columns='country', values='rgdpo')

df_pivot = np.log(df_pivot)

india_cycle, india_trend = hpfilter(df_pivot['India'], lamb=1600)
japan_cycle, japan_trend = hpfilter(df_pivot['Japan'], lamb=1600)

df_pivot['India_Cycle'] = india_cycle
df_pivot['India_Trend'] = india_trend
df_pivot['Japan_Cycle'] = japan_cycle
df_pivot['Japan_Trend'] = japan_trend

india_cycle_std = df_pivot['India_Cycle'].std()
japan_cycle_std = df_pivot['Japan_Cycle'].std()

correlation = df_pivot[['India_Cycle', 'Japan_Cycle']].corr().iloc[0, 1]

print(f"India's Cycle Standard Deviation: {india_cycle_std}")
print(f"Japan's Cycle Standard Deviation: {japan_cycle_std}")
print(f"Correlation between India and Japan Cycles: {correlation}")

plt.figure(figsize=(14, 7))
plt.plot(df_pivot.index, df_pivot['India_Cycle'], label='India Cycle')
plt.plot(df_pivot.index, df_pivot['Japan_Cycle'], label='Japan Cycle')
plt.title('HP Filtered Cycles of India and Japan')
plt.xlabel('Year')
plt.ylabel('Cycle Component')
plt.legend()
plt.show()
```

```python
import pandas as pd

pwt = pd.read_csv('/content/pwt1001.csv')

oecd_countries = [
    "Australia", "Austria", "Belgium", "Canada", "Denmark", "Finland", "France",
    "Germany", "Greece", "Iceland", "Ireland", "Italy", "Japan", "Netherlands",
    "New Zealand", "Norway", "Portugal", "Spain", "Sweden", "Switzerland",
    "United Kingdom", "United States"
]

columns = ['country', 'year', 'rgdpo', 'rkna', 'emp', 'labsh']

df = pwt[(pwt['country'].isin(oecd_countries)) & (pwt['year'] >= 1990) & (pwt['year'] <= 2019)][columns]

results = []

for country in oecd_countries:
    country_data = df[df['country'] == country].dropna()
    country_data['y_growth'] = country_data['rgdpo'].pct_change()
    country_data['k_growth'] = country_data['rkna'].pct_change()
    country_data['l_growth'] = country_data['emp'].pct_change()

    country_data['tfp_growth'] = country_data['y_growth'] - country_data['labsh'] * country_data['l_growth'] - (1 - country_data['labsh']) * country_data['k_growth']
    country_data['capital_deepening'] = (1 - country_data['labsh']) * country_data['k_growth']

    avg_growth_rate = country_data['y_growth'].mean() * 100
    avg_tfp_growth = country_data['tfp_growth'].mean() * 100
    avg_capital_deepening = country_data['capital_deepening'].mean() * 100
    tfp_share = avg_tfp_growth / avg_growth_rate
    capital_share = avg_capital_deepening / avg_growth_rate

    results.append([country, avg_growth_rate, avg_tfp_growth, avg_capital_deepening, tfp_share, capital_share])

results_df = pd.DataFrame(results, columns=['Country', 'Growth Rate', 'TFP Growth', 'Capital Deepening', 'TFP Share', 'Capital Share'])

print(results_df)

averages = results_df.mean(numeric_only=True) # Specify numeric_only to calculate mean for numerical columns
print("\nAverages:")
print(averages)
```
